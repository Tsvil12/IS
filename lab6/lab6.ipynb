Напишите класс Balance для описания весов с двумя чашами. На левую и правую чашу объекта будут добавляться грузы с различным весом, ваша задача определить положение чаш.

Метод add_right принимает целое число — вес, положенный на правую чашу весов, add_left — на левую чашу. Метод result должен возвращать символ =, если вес на чашах одинаковый, R — если перевесила правая, L — если перевесила левая.

```python
class Balance:
    def __init__(self):
        self.left = 0
        self.right = 0
    
    def add_left(self, weight):
        self.left += weight
    
    def add_right(self, weight):
        self.right += weight
    
    def result(self):
        if self.left > self.right:
            return 'L'
        elif self.right > self.left:
            return 'R'
        else:
            return '='

balance = Balance()
balance.add_left(10)
balance.add_right(5)
print(balance.result())
balance.add_left(3)
print(balance.result())
balance.add_right(8)
print(balance.result())
```

Напишите класс Selector. Экземпляр этого класса при инициализации получает список чисел. Вызов метода get_odds возвращает нечётные числа из первоначального списка, вызов get_evens — чётные.

Числа должны идти в том же порядке, в котором они были в изначальном списке.

```python
class Selector:
    def __init__(self, numbers):
        self.numbers = numbers
    
    def get_odds(self):
        return [x for x in self.numbers if x % 2 != 0]
    
    def get_evens(self):
        return [x for x in self.numbers if x % 2 == 0]

selector = Selector([1, 2, 3, 4, 5, 6, 7, 8, 9])
print(selector.get_odds())
print(selector.get_evens())
```

Определите класс Point. При инициализации экземпляру передаются координаты x и y.

При сравнении двух экземпляров оператор == должен возвращать True, если координаты точек равны, и False — если нет.

При сравнении оператором != должно возвращаться True, если координаты точек не равны, и False — если равны.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

p1 = Point(1, 2)
p2 = Point(1, 2)
p3 = Point(3, 4)

print(p1 == p2)
print(p1 == p3)
print(p1 != p2)
print(p1 != p3)
```

```python
class ReversedList:
    def __init__(self, lst):
        self.lst = lst
    
    def __len__(self):
        return len(self.lst)
    
    def __getitem__(self, index):
        return self.lst[-index - 1]

rl = ReversedList([10, 20, 30, 40, 50])
print(len(rl))
print(rl[0])
print(rl[1])
print(rl[2])
print(rl[3])
print(rl[4])
```

Периодически в некоторых задачах возникает надобность в разреженных массивах. Разреженный массив характеризуется тем, что подавляющее большинство значений в нём равны нулю, поэтому можно хранить только ненулевые значения. Это позволяет создавать разреженные массивы очень большого размера без лишних затрат памяти. Вам необходимо реализовать класс разреженного массива SparseArray.

arr = SparseArray() — создание пустого разреженного массива.

arr[i] = value — записать значение в массив.

arr[i] — прочитать значение из массива.

```python
class SparseArray:
    def __init__(self):
        self.data = {}
    
    def __setitem__(self, index, value):
        if value != 0:
            self.data[index] = value
        elif index in self.data:
            del self.data[index]
    
    def __getitem__(self, index):
        return self.data.get(index, 0)

arr = SparseArray()
arr[10] = 100
arr[20] = 200
arr[30] = 0

print(arr[10])
print(arr[20])
print(arr[30])
print(arr[1000])
```

Реализуйте класс Polynomial для вычисления значений многочленов при заданных x, а также для сложения многочленов.

poly = Polynomial(coefficients) — создать новый многочлен с коэффициентами coefficients.

coefficients[0] — свободный член, coefficients[1] — множитель при x¹, coefficients[2] — множитель при x², и так далее.

y = poly(x) — вычислить значение многочлена в точке x.

poly_sum = poly1 + poly2 — результат сложения двух многочленов должен быть экземпляром Polynomial и содержать коэффициенты, которые получаются в результате сложения коэффициентов poly1 и poly2 при одинаковых степенях x.

Примечание: модуль numpy в этой задаче использовать нельзя.

```python
class Polynomial:
    def __init__(self, coefficients):
        self.coefficients = coefficients
    
    def __call__(self, x):
        result = 0
        for power, coef in enumerate(self.coefficients):
            result += coef * (x ** power)
        return result
    
    def __add__(self, other):
        max_len = max(len(self.coefficients), len(other.coefficients))
        new_coeffs = []
        for i in range(max_len):
            a = self.coefficients[i] if i < len(self.coefficients) else 0
            b = other.coefficients[i] if i < len(other.coefficients) else 0
            new_coeffs.append(a + b)
        return Polynomial(new_coeffs)

poly1 = Polynomial([1, 2, 3])
print(poly1(2))

poly2 = Polynomial([0, 1, 0, 1])
poly_sum = poly1 + poly2
print(poly_sum.coefficients)
print(poly_sum(2))
```

Реализуйте класс Queue, моделирующий работу очереди (принцип FIFO (First In First Out)). Очередь может быть инициализирована различным числом элементов

```python
class Queue:
    def __init__(self, *args):
        self.items = list(args)
    
    def add(self, item):
        self.items.append(item)
    
    def pop(self):
        if not self.items:
            return None
        return self.items.pop(0)
    
    def __len__(self):
        return len(self.items)
    
    def __str__(self):
        return str(self.items)

q = Queue(1, 2, 3)
print(q)
q.add(4)
print(q)
print(q.pop())
print(q)
print(len(q))
```

Реализуйте несколько классов, вычисляющих суммы следующих последовательностей натуральных чисел от 1 до N:

Сделайте это с помощью наследования.
Базовым классом будет Summat, который реализует методы transform(n) и sum(N).

Метод transform(n) выполняет преобразование элемента последовательности. Для класса Summat это будет тождественное преобразование, т.е. transform(n) == n.
Метод transform(n) будет переопределяться в производных классах.

Метод sum(N) должен вычислять сумму значений transform(n) для всех значений n от 1 до N.

Производные классы должны переопределять метод transform(n):

SquareSummat: transform(n) == n ** 2

CubeSummat: transform(n) == n ** 3

Метод sum(N) переопределять не нужно

```python
class Summat:
    def transform(self, n):
        return n
    
    def sum(self, N):
        total = 0
        for i in range(1, N + 1):
            total += self.transform(i)
        return total

class SquareSummat(Summat):
    def transform(self, n):
        return n ** 2

class CubeSummat(Summat):
    def transform(self, n):
        return n ** 3

s = Summat()
print(s.sum(5))

sq = SquareSummat()
print(sq.sum(5))

cb = CubeSummat()
print(cb.sum(5))
```

PowerSummator должен наследоваться от класса Summator из предыдущего задания.

Помимо метода transform, для класса PowerSummator также нужно переопределить конструктор. Конструктор класса PowerSummator должен принимать один параметр — степень b, в которую нужно возводить элементы последовательности.

В классах SquareSummator и CubeSummator должен быть переопределён только конструктор; переопределять другие методы нельзя

```python
class Summator:
    def transform(self, n):
        return n
    
    def sum(self, N):
        total = 0
        for i in range(1, N + 1):
            total += self.transform(i)
        return total

class PowerSummator(Summator):
    def __init__(self, b):
        self.b = b
    
    def transform(self, n):
        return n ** self.b

class SquareSummator(PowerSummator):
    def __init__(self):
        super().__init__(2)

class CubeSummator(PowerSummator):
    def __init__(self):
        super().__init__(3)

ps = PowerSummator(2.5)
print(ps.sum(4))

sq = SquareSummator()
print(sq.sum(5))

cb = CubeSummator()
print(cb.sum(5))
```

Реализуйте программу, имитирующую работу почтовых серверов и клиентов. 
Клиенты должны иметь возможность принимать и отправлять письма, которые хранятся на серверах.
 Вам предстоит спроектировать программу самостоятельно. Чтобы задать направление мысли, давайте обсудим примерный интерфейс почтового клиента. 
client = MailClient(server, user) — подключение к учётной записи пользователя user на сервере server. 
client.receive_mail() — получение почты. Когда почта получена, она удаляется с сервера. Что конкретно возвращает этот метод, решите самостоятельно. 
client.send_mail(server1, user1, message) — послать сообщение message пользователю user1 на сервере server1.
 Ваша программа должна быть не просто набором классов, но и предоставлять готовый интерфейс, по которому преподаватель будет оценивать функциональность и корректность вашей программы. 
Программа должна позволять из своего интерфейса определять список серверов, на которые можно отправлять письма. При попытке отправить письмо на сервер, которого нет в списке, программа должна сообщать пользователю, что не может этого сделать

```python
class Server:
    def __init__(self, name):
        self.name = name
        self.messages = {}
    
    def add_user(self, user):
        if user not in self.messages:
            self.messages[user] = []
    
    def send_to_server(self, to_user, msg):
        if to_user in self.messages:
            self.messages[to_user].append(msg)
            return True
        return False
    
    def get_mail(self, user):
        if self.messages[user]:
            return self.messages[user].pop(0)
        return None

class Client:
    servers = []
    
    def __init__(self, server, name):
        self.server = server
        self.name = name
        server.add_user(name)
    
    def send(self, to_server, to_name, text):
        if to_server not in Client.servers:
            print("Такого сервера нет")
            return
        to_server.send_to_server(to_name, f"От {self.name}: {text}")
        print("Отправлено!")
    
    def receive(self):
        msg = self.server.get_mail(self.name)
        if msg:
            print(msg)
        else:
            print("Нет писем")

s1 = Server("Mail.ru")
s2 = Server("Yandex.ru")

Client.servers = [s1, s2]

alice = Client(s1, "Алиса")
bob = Client(s2, "Боб")

alice.send(s2, "Боб", "Привет!")
bob.receive()
```

# №15
 Нужно написать программу для преобразования списка чисел в соответствии с некоторым набором правил. 
В начале своей работы программа считывает числа. Числа записаны в строчку через пробел.
 Затем к списку применяются одна из команд make_negative, square, strange_command.
 Каждая из команд состоит из условия, с помощью которого выбираются числа, и преобразования, которому выбранные числа подвергаются. 
Так, make_negative применяется только к положительным числам и умножает их на -1, square применяется ко всем числам и возводит их в квадрат, strange_command действует только на числа, делящиеся на 5, и увеличивает их на 1. 
Постарайтесь спроектировать программу так, чтобы при добавлении новых подобных преобразований, так же определяющихся именем, условием и самим преобразованием, количество изменений в уже написанном коде было минимальным.


```python
class Command:
    def __init__(self, condition, transform):
        self.condition = condition
        self.transform = transform
    
    def apply(self, num):
        if self.condition(num):
            return self.transform(num)
        return num

def make_negative():
    return Command(lambda x: x > 0, lambda x: x * -1)

def square():
    return Command(lambda x: True, lambda x: x ** 2)

def strange_command():
    return Command(lambda x: x % 5 == 0, lambda x: x + 1)

numbers = list(map(int, input("Введите числа: ").split()))

commands = [make_negative(), square(), strange_command()]

for cmd in commands:
    numbers = [cmd.apply(n) for n in numbers]

print(numbers)
```

Вам нужно написать программу для планирования выступлений на конференциях. Её функциональность должна быть следующей: можно создавать выступление, задавать его тему, время его начала и длительность. Программа должна следить за тем, чтобы доклады не перекрывались во времени и предупреждать пользователя, если это произошло, выводить суммарное время доклада, время самого продолжительного перерыва между докладами и так далее. Надо реализовать два класса: «доклад» и «конференция». Как распределить между ними функциональность, решать Вам. Программа должна иметь хороший интерфейс с подсказками, что нужно вводить

```python
class Talk:
    def __init__(self, title, start_hour, start_minute, duration):
        self.title = title
        self.start = start_hour * 60 + start_minute
        self.end = self.start + duration
        self.duration = duration
    
    def __str__(self):
        h = self.start // 60
        m = self.start % 60
        return f"{self.title} - {h}:{m:02d} ({self.duration} мин)"

class Conference:
    def __init__(self):
        self.talks = []
    
    def add_talk(self, talk):
        for t in self.talks:
            if not (talk.end <= t.start or talk.start >= t.end):
                print(f"Ошибка! {talk.title} пересекается с {t.title}")
                return False
        self.talks.append(talk)
        self.talks.sort(key=lambda x: x.start)
        print(f"Доклад '{talk.title}' добавлен")
        return True
    
    def show_schedule(self):
        print("\nРасписание:")
        for talk in self.talks:
            print(talk)
    
    def total_time(self):
        total = sum(t.duration for t in self.talks)
        print(f"Общее время докладов: {total} минут")
    
    def max_break(self):
        if len(self.talks) < 2:
            print("Недостаточно докладов")
            return
        max_break = 0
        for i in range(1, len(self.talks)):
            gap = self.talks[i].start - self.talks[i-1].end
            if gap > max_break:
                max_break = gap
        print(f"Самый длинный перерыв: {max_break} минут")

conf = Conference()

while True:
    print("\n1. Добавить доклад")
    print("2. Показать расписание")
    print("3. Общее время")
    print("4. Самый длинный перерыв")
    print("5. Выйти")
    
    choice = input("Выберите действие: ")
    
    if choice == "1":
        title = input("Тема: ")
        h = int(input("Час начала: "))
        m = int(input("Минута начала: "))
        d = int(input("Длительность (мин): "))
        conf.add_talk(Talk(title, h, m, d))
    elif choice == "2":
        conf.show_schedule()
    elif choice == "3":
        conf.total_time()
    elif choice == "4":
        conf.max_break()
    elif choice == "5":
        break
```

Имена файлов и директорий указываются относительно общей корневой директории. В качестве разделителя между компонентами пути в файловой системе удобнее всего использовать символ "/".

Старайтесь проектировать программу так, чтобы у вас не было общего списка или словаря всех файлов в файловой системе, и директории хранили свои поддиректории и файлы, как в настоящих файловых системах.

```python
class Node:
    def __init__(self, name):
        self.name = name
        self.children = {}
        self.files = {}

class FileSystem:
    def __init__(self):
        self.root = Node("")
    
    def _get_node(self, path):
        if not path:
            return self.root
        parts = path.split("/")
        current = self.root
        for p in parts:
            if p in current.children:
                current = current.children[p]
            else:
                return None
        return current
    
    def mkdir(self, path):
        parts = path.split("/")
        current = self.root
        for p in parts:
            if p not in current.children:
                current.children[p] = Node(p)
            current = current.children[p]
        print(f"Директория '{path}' создана")
    
    def ls(self, path):
        node = self._get_node(path)
        if node:
            items = list(node.children.keys()) + list(node.files.keys())
            if items:
                print(" ".join(items))
            else:
                print("(пусто)")
        else:
            print("Директория не найдена")
    
    def write(self, path, content):
        parts = path.split("/")
        filename = parts[-1]
        dir_path = "/".join(parts[:-1]) if len(parts) > 1 else ""
        
        parent = self._get_node(dir_path)
        if parent:
            parent.files[filename] = content
            print(f"Файл '{path}' записан")
        else:
            print("Путь не найден")
    
    def read(self, path):
        parts = path.split("/")
        filename = parts[-1]
        dir_path = "/".join(parts[:-1]) if len(parts) > 1 else ""
        
        parent = self._get_node(dir_path)
        if parent and filename in parent.files:
            print(parent.files[filename])
        else:
            print("Файл не найден")

fs = FileSystem()

while True:
    print("\nКоманды: mkdir, ls, write, read, exit")
    cmd = input("> ").strip().split(maxsplit=1)
    
    if not cmd:
        continue
    
    if cmd[0] == "exit":
        break
    elif cmd[0] == "mkdir" and len(cmd) > 1:
        fs.mkdir(cmd[1])
    elif cmd[0] == "ls" and len(cmd) > 1:
        fs.ls(cmd[1])
    elif cmd[0] == "ls":
        fs.ls("")
    elif cmd[0] == "write" and len(cmd) > 1:
        parts = cmd[1].split(maxsplit=1)
        if len(parts) == 2:
            fs.write(parts[0], parts[1])
        else:
            print("Использование: write путь содержимое")
    elif cmd[0] == "read" and len(cmd) > 1:
        fs.read(cmd[1])
    else:
        print("Неверная команда")
```
