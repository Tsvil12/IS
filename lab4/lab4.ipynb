# №5
Напишите функцию number_to_words(n), которая будет возвращать переданное в качестве параметра n число словами.

```python
def number_to_words(n):
    units = ['', 'один', 'два', 'три', 'четыре', 'пять', 'шесть', 'семь', 'восемь', 'девять']
    teens = ['десять', 'одиннадцать', 'двенадцать', 'тринадцать', 'четырнадцать', 'пятнадцать', 'шестнадцать', 'семнадцать', 'восемнадцать', 'девятнадцать']
    tens = ['', '', 'двадцать', 'тридцать', 'сорок', 'пятьдесят', 'шестьдесят', 'семьдесят', 'восемьдесят', 'девяносто']
    
    if 1 <= n <= 9:
        return units[n]
    elif 10 <= n <= 19:
        return teens[n - 10]
    elif 20 <= n <= 99:
        ten = n // 10
        unit = n % 10
        if unit == 0:
            return tens[ten]
        else:
            return tens[ten] + ' ' + units[unit]

n = int(input("Введите число от 1 до 99: "))
print(number_to_words(n))
```

# №6
Правильной скобочной последовательностью называется строка, состоящая только из символов «скобки» (открывающих "(" и закрывающих ")"), где каждой закрывающей скобке найдётся соответствующая открывающая. Например, () и (()()) — правильные последовательности, а (()(() или )() — нет.

Напишите функцию bracket_check(test_string), которая проверяет, является ли поступившая на вход строка правильной скобочной последовательностью. Если да, она должна печатать YES, в противном случае — NO. Обратите внимание, что пустая строка также является правильной скобочной последовательностью

```python
def bracket_check(test_string):
    balance = 0
    for char in test_string:
        if char == '(':
            balance += 1
        elif char == ')':
            balance -= 1
            if balance < 0:
                print("NO")
                return
    if balance == 0:
        print("YES")
    else:
        print("NO")

test_string = input("Введите строку из скобок: ")
bracket_check(test_string)
```

# №7
Палиндром — это строка (или слово), которая читается одинаково в обоих направлениях, например, «12321» или известный пример Афанасия Фета «А роза упала на лапу Азора».

Напишите функцию palindrome(s), которая принимает в качества параметра строку и определяет, является ли она палиндромом. Функция должна возвращать строку «Палиндром» или «Не палиндром» соответственно.

```python
def palindrome(s):
    s_clean = s.replace(" ", "").lower()
    if s_clean == s_clean[::-1]:
        return "Палиндром"
    else:
        return "Не палиндром"

s = input("Введите строку: ")
print(palindrome(s))
```

Напишите функцию tic_tac_toe(field), которая принимает на вход список списков 3x3 — поле игры в крестики-нолики.

На этом поле отражен игровой момент: незанятой (пустой) клетке соответствует ‘-‘, крестикам и ноликам соответственно ‘x’ и ‘0’.

Ваша функция должна попытаться определить, есть ли победитель в данном игровом моменте и вывести фразу x win, если победили «крестики» или 0 win, если победили «нолики».

В случае, когда однозначно определить победителя не представляется возможным необходимо вывести слово draw.

```python
def tic_tac_toe(field):
    for row in field:
        if row.count(row[0]) == 3 and row[0] != '*':
            return f"{row[0]} win"

    for col in range(3):
        if field[0][col] == field[1][col] == field[2][col] != '*':
            return f"{field[0][col]} win"

    if field[0][0] == field[1][1] == field[2][2] != '*':
        return f"{field[0][0]} win"
    if field[0][2] == field[1][1] == field[2][0] != '*':
        return f"{field[2][0]} win"

    return "draw"

game_field = [
    ['x', 'x', 'x'],
    ['0', '*', '0'],
    ['*', '0', '*']
]

result = tic_tac_toe(game_field)
print(result)
```

Из-за плохой связи на мобильный телефон Василия приходят одни и те же смски по несколько раз. Он попросил вас написать небольшую функцию, которая будет работать почти как обычный print, но не будет печатать сообщение, если предыдущее сообщение было таким же.

Чтобы телефон эти изменения в распечатке смсок принял, ваш друг настроил его так, что вам достаточно написать функцию print_without_duplicates(message). Напишите её.

```python
last_message = None

def print_without_duplicates(message):
    global last_message
    if message != last_message:
        print(message)
        last_message = message

print("Введите сообщения. Для завершения введите 'выход'")
while True:
    msg = input("Сообщение: ")
    if msg == "выход":
        break
    print_without_duplicates(msg)
```

# №10
Алиса хочет устроить вечеринку по поводу ее дня рождения. Она хочет пригласить друзей так, чтобы друзья были между собой знакомы.

Помогите Алисе автоматизировать процесс написания приглашений. Напишите три функции: add_friends(), are_friends() и print_friends().

Функция add_friends(name_of_person, list_of_friends) принимает на вход два аргумента: имя человека и список имен его друзей.

Функция are_friends(name_of_person1, name_of_person2) принимает на вход два аргумента: имя первого человека и имя второго человека и возвращает True или False, в зависимости от того, считает ли первый человек второго своим другом.

А функция print_friends(name_of_person) принимает на вход имя человека и печатает список его друзей через пробел в алфавитном порядке.

```python
friends_db = {}

def add_friends(name_of_person, list_of_friends):
    if name_of_person not in friends_db:
        friends_db[name_of_person] = set()
    for friend in list_of_friends:
        friends_db[name_of_person].add(friend)

def are_friends(name_of_person1, name_of_person2):
    if name_of_person1 not in friends_db:
        return False
    return name_of_person2 in friends_db[name_of_person1]

def print_friends(name_of_person):
    if name_of_person in friends_db:
        print(" ".join(sorted(friends_db[name_of_person])))
    else:
        print("У этого человека нет друзей")

print("ДОБАВЛЕНИЕ ДРУЗЕЙ")
print("Введите '0' для завершения")
while True:
    name = input("Имя человека: ")
    if name == "0":
        break
    friends_input = input("Друзья через пробел: ").split()
    add_friends(name, friends_input)

print("\nПРОВЕРКА ДРУЖБЫ")
print("Введите '0' для завершения")
while True:
    p1 = input("Первый человек: ")
    if p1 == "0":
        break
    p2 = input("Второй человек: ")
    if are_friends(p1, p2):
        print(f"ДА, {p1} считает {p2} другом")
    else:
        print(f"НЕТ, {p1} не считает {p2} другом")

print("\nСПИСОК ДРУЗЕЙ")
print("Введите '0' для завершения")
while True:
    name = input("Человек: ")
    if name == "0":
        break
    print_friends(name)

print("\nПрограмма завершена")
```

Формат ввода
one = 5

Формат вывода

Примечания
Напомним, как соотносятся арабские и римские числа:

1 – I

5 – V

10 – X

50 – L

100 – C

500 – D

1000 – M

Не забудьте про методику сокращения записи в римской системе счисления: например, 4 – IV.

```python
def to_roman(num):
    values = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
    symbols = ["M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]
    
    result = ""
    for i in range(len(values)):
        while num >= values[i]:
            result += symbols[i]
            num -= values[i]
    return result

def roman():
    global one, two, three
    three = one + two
    print(f"{to_roman(one)} + {to_roman(two)} = {to_roman(three)}")

one = 5
two = 4
roman()
```

# №12
Приведите пример, когда value = value + addition и value += addition дают одинаковый результат и когда разный. В чем будут отличия, и как их пронаблюдать? Напишите код, который их продемонстрирует. В коде обязательно должны присутствовать комментарии, поясняющие работу программы.

```python
# Для неизменяемых типов (int, str, tuple) оба оператора работают одинаково потому что создаётся новый объект

print("НЕИЗМЕНЯЕМЫЕ ТИПЫ (int)")
a = 5
b = a
a = a + 3
print(f"a = a + 3: a={a}, b={b}")  # b остался 5

a = 5
b = a
a += 3
print(f"a += 3: a={a}, b={b}")      # b остался 5
# Результат одинаковый

print("\nНЕИЗМЕНЯЕМЫЕ ТИПЫ (str)")
s = "Hello"
t = s
s = s + " World"
print(f"s = s + ' World': s={s}, t={t}")

s = "Hello"
t = s
s += " World"
print(f"s += ' World': s={s}, t={t}")
# Результат одинаковый

# Для изменяемых типов (list, dict, set) поведение РАЗНОЕ
# += изменяет исходный объект, а = + создаёт новый

print("\nИЗМЕНЯЕМЫЕ ТИПЫ (list)")
lst1 = [1, 2, 3]
lst2 = lst1
lst1 = lst1 + [4, 5]
print(f"lst1 = lst1 + [4,5]: lst1={lst1}, lst2={lst2}")
# lst2 остался старым

lst1 = [1, 2, 3]
lst2 = lst1
lst1 += [4, 5]
print(f"lst1 += [4,5]: lst1={lst1}, lst2={lst2}")
# lst2 тоже изменился! Потому что += изменяет исходный список

print("\nВЫВОД")
print("Для неизменяемых типов: оба оператора создают новый объект")
print("Для изменяемых типов: += изменяет объект, = + создаёт новый")
```

Напишите пример, который продемонстрирует отличия между методом sort() списка и функцией sorted(arr). На основе вашего кода должно быть возможно показать, какой из вариантов меняет список, а какой – создает новый, отсортированный список. В коде обязательно должны присутствовать комментарии, поясняющие работу программы.

```python
# Метод sort() - сортирует список на месте, изменяя исходный
# Функция sorted() - создаёт новый отсортированный список, исходный не меняет

print("=== Метод list.sort() ===")
numbers1 = [3, 1, 4, 1, 5, 9, 2]
print(f"Исходный список: {numbers1}")

numbers1.sort()
print(f"После sort(): {numbers1}")
print("Список был изменён (отсортирован на месте)\n")

print("=== Функция sorted() ===")
numbers2 = [3, 1, 4, 1, 5, 9, 2]
print(f"Исходный список: {numbers2}")

sorted_numbers = sorted(numbers2)
print(f"Результат sorted(): {sorted_numbers}")
print(f"Исходный список после sorted(): {numbers2}")
print("Исходный список НЕ ИЗМЕНИЛСЯ\n")

print("=== Демонстрация с разными типами ===")
words = ["banana", "apple", "cherry", "date"]
print(f"Исходный список слов: {words}")

# sort() работает с любыми сравнимыми элементами
words.sort(reverse=True)
print(f"После sort(reverse=True): {words}")

# sorted() можно использовать для любых итерируемых объектов
word_tuple = ("zebra", "apple", "monkey", "cat")
print(f"Кортеж: {word_tuple}")
sorted_from_tuple = sorted(word_tuple)
print(f"sorted() из кортежа: {sorted_from_tuple}")
print(f"Исходный кортеж не изменился: {word_tuple}")

print("\n=== ИТОГ ===")
print("sort() - изменяет исходный список, возвращает None")
print("sorted() - не изменяет исходный, возвращает новый отсортированный список")
```

```python
# Исходный код с ошибкой:
# def continue_fibonacci_sequence(sequence, n):
#     for i in range(n):
#     next_element = sequence[-1] + sequence[-2]
#     sequence = sequence + [next_element]

# Ошибки:
# 1. Отсутствует отступ у тела цикла (строка с next_element и sequence должны быть внутри for)
# 2. Внутри цикла создаётся НОВЫЙ список sequence, а не изменяется исходный
#    Исходный список вне функции останется неизменным

# Исправленный код:

def continue_fibonacci_sequence(sequence, n):
    # Добавляем n новых элементов в последовательность
    for i in range(n):
        next_element = sequence[-1] + sequence[-2]  # сумма двух последних
        sequence.append(next_element)  # изменяем исходный список через append

# Демонстрация работы
fib = [1, 1]
print("Исходная последовательность:", fib)
continue_fibonacci_sequence(fib, 10)
print("После добавления 10 элементов:", fib)

# Альтернативный вариант (если нужно вернуть новый список, не меняя исходный):
def continue_fibonacci_sequence_new(sequence, n):
    result = sequence.copy()  # создаём копию исходного списка
    for i in range(n):
        result.append(result[-1] + result[-2])
    return result

fib2 = [1, 1]
print("\nИсходная последовательность (для нового варианта):", fib2)
new_fib = continue_fibonacci_sequence_new(fib2, 10)
print("Исходная осталась неизменной:", fib2)
print("Новый список:", new_fib)
```

Посмотрите на функцию mirror(arr), которая должна список длины

N
2N, первая половина которого равна исходному списку, а вторая половина равна перевёрнутому («отражённому» относительно конца):

```python
# Исходный код с ошибкой:
# def mirror(arr):
#     mirrored_part = arr.reverse()
#     arr = arr + mirrored_part

# Ошибки:
# 1. метод reverse() изменяет список на месте и возвращает None
# 2. mirrored_part получает значение None, а не перевёрнутый список
# 3. arr = ... создаёт новую локальную переменную, исходный список не меняется

# Исправленный код:

def mirror(arr):
    # Создаём копию списка и переворачиваем её
    mirrored_part = arr.copy()
    mirrored_part.reverse()
    # Добавляем перевёрнутую часть в исходный список
    arr.extend(mirrored_part)

# Демонстрация работы
my_list = [1, 2, 3, 4]
print("Исходный список:", my_list)
mirror(my_list)
print("После mirror():", my_list)
# Ожидаемый результат: [1, 2, 3, 4, 4, 3, 2, 1]

# Альтернативный вариант (короткий, без .extend):
def mirror2(arr):
    arr += arr[::-1]

my_list2 = [1, 2, 3, 4]
print("\nИсходный список (вариант 2):", my_list2)
mirror2(my_list2)
print("После mirror2():", my_list2)
```

# №19
Напишите функцию transpose(matrix), которая принимает матрицу 
N
×
M
N×M (т.е. список из 
N
N списков-строк по 
M
M элементов каждый) и транспонирует её, т.е. превращает строки в столбцы и наоборот.

i
i-ая строка при этом превращается в 
i
i-й столбец.

```python
def transpose(matrix):
    if not matrix:
        return []
    
    rows = len(matrix)
    cols = len(matrix[0])
    
    result = []
    for j in range(cols):
        new_row = []
        for i in range(rows):
            new_row.append(matrix[i][j])
        result.append(new_row)
    
    return result

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Исходная матрица:")
for row in matrix:
    print(row)

transposed = transpose(matrix)

print("\nТранспонированная матрица:")
for row in transposed:
    print(row)

def transpose_zip(matrix):
    return [list(row) for row in zip(*matrix)]

print("\nТранспонированная матрица (через zip):")
for row in transpose_zip(matrix):
    print(row)
```

Функция должна называться swap(first, second). Первый список должен получить содержимое второго списка, и наоборот.

```python
def swap2(first, second):
    temp = first[:]
    first[:] = second[:]
    second[:] = temp[:]

thoughts3 = ["a", "b", "c"]
thoughts4 = ["x", "y", "z"]

print("\n" + "="*40)
print("До обмена (вариант 2):")
print("Первый:", thoughts3)
print("Второй:", thoughts4)

swap2(thoughts3, thoughts4)

print("После обмена:")
print("Первый:", thoughts3)
print("Второй:", thoughts4)
```

# №21
Напишите функцию defractalize(fractal), которая принимает в качестве аргумента фрактальный список и удаляет из него все самоподобные части фрактала (т.е. части, эквивалентные самому фракталу).

Фрактальный список в этой задаче может содержать в качестве элементов любые числа и сам себя (в любом числе копий).

```python
def defractalize(fractal):
    i = 0
    while i < len(fractal):
        if fractal[i] is fractal:
            fractal.pop(i)
        else:
            i += 1

fractal = [1, 2, [3, 4], [1, 2, [3, 4]]]
print("Исходный список:", fractal)

f = [1, 2, 3]
f.append(f)
print("\nФрактальный список:", f)

defractalize(f)
print("После defractalize():", f)
```

Это позволяет программе не зациклиться, пытаясь напечатать список, который за счет фрактальной структуры имеет бесконечную глубину.

Например, текстовое представление списка, состоящего из нуля, самого себя и двойки будет выглядеть так:

[0, [...], 2]

Напишите функцию fractal_print(obj), которая также выводит фрактал на экран, но печатает не один, а два уровня вложенности:

[0, [0, [...], 2], 2]

Мы рассматриваем только простые фрактальные списки, состоящие из обычных несчастных объектов (числа, строки, логические величины) или самих себя.

```python
def fractal_print(obj):
    def helper(item, depth):
        if depth >= 2:
            return "[...]"
        
        if isinstance(item, list):
            result = "["
            for i, elem in enumerate(item):
                if elem is obj and depth == 0:
                    result += helper(elem, depth + 1)
                elif isinstance(elem, list):
                    result += helper(elem, depth + 1)
                else:
                    result += str(elem)
                if i < len(item) - 1:
                    result += ", "
            result += "]"
            return result
        else:
            return str(item)
    
    print(helper(obj, 0))

f = [0, None, 2]
f[1] = f
print("Обычный print:", f)
print("fractal_print:", end=" ")
fractal_print(f)
```

k-ой частичной суммой списка называется сумма первых k элементов списка. Напишите функцию partial_sums, которая принимает неограниченное число аргументов, а возвращает список частичных сумм этих элементов: на нулевой позиции — 0, на первой позиции — первое число, на второй — сумму первого и второго чисел, затем — сумму первого, второго и третьего и т.д.

Обратите внимание, что функция должна принимать не список, а именно неограниченное число аргументов.

```python
def partial_sums(*args):
    result = [0]
    current_sum = 0
    for num in args:
        current_sum += num
        result.append(current_sum)
    return result

print(partial_sums(1, 2, 3, 4, 5))
print(partial_sums(10, 20, 30))
print(partial_sums())
```

# №25
Сделайте функцию solve(*coefficients), которая умеет решать уравнения степени не выше второй (квадратные и линейные).

Если у функции три аргумента, их надо трактовать как a, b и c в уравнении ax² + bx + c = 0.

Если два — как коэффициенты b и c в уравнении bx + c = 0.

Если один — как коэффициент c в уравнении c = 0.

Если список коэффициентов пуст или коэффициентов больше трёх, то функция должна вернуть None.

Возвращаемое значение у функции ровно одно, а корней может быть от нуля до двух (есть еще один случай: a = b = c = 0, при котором решением является любое число; про него ниже), поэтому сделать так, чтобы эта функция возвращала просто значение корня, не получится.

Вместо этого вы должны вернуть из функции список, содержащий значения всех корней уравнения (одно, два или ни одного). Если корнем является любое значение x, верните вместо списка чисел — список, содержащий одну строку "all" (["all"])

Не забудьте рассмотреть случаи, когда квадратное уравнение выражается до линейного (a = 0).

Также бывает, что линейное уравнение выражается еще дальше — до уравнения, в котором x не участвует вовсе (a = b = 0). В этом случае мы получаем уравнение c = 0, которое, в зависимости от значения c, либо не имеет корней, либо корнем является любой x.

```python
import math

def solve(*coefficients):
    if len(coefficients) == 0 or len(coefficients) > 3:
        return None
    
    if len(coefficients) == 1:
        c = coefficients[0]
        if c == 0:
            return ["all"]
        else:
            return []
    
    elif len(coefficients) == 2:
        b, c = coefficients
        if b == 0:
            if c == 0:
                return ["all"]
            else:
                return []
        else:
            x = -c / b
            return [x]
    
    else:  # len(coefficients) == 3
        a, b, c = coefficients
        
        if a == 0:
            return solve(b, c)
        
        d = b**2 - 4*a*c
        
        if d < 0:
            return []
        elif d == 0:
            x = -b / (2*a)
            return [x]
        else:
            x1 = (-b - math.sqrt(d)) / (2*a)
            x2 = (-b + math.sqrt(d)) / (2*a)
            return [x1, x2]

print(solve(1, -3, 2))
print(solve(1, 2, 1))  
print(solve(1, 0, 1))  
print(solve(2, 4))
print(solve(0, 5))      
print(solve(0, 0))     
print(solve(5))      
print(solve(0))
print(solve())     
```
